```python
# normalize features
num_vars = [
    'PRICE_RATE',
    'CATALOG_PRICE',
    'DISPPERIOD',
    'VALIDPERIOD',
]
bin_vars = [
    'USABLE_DATE_MON',
    'USABLE_DATE_TUE',
    'USABLE_DATE_WED',
    'USABLE_DATE_THU',
    'USABLE_DATE_FRI',
    'USABLE_DATE_SAT',
    'USABLE_DATE_SUN',
    'USABLE_DATE_HOLIDAY',
    'USABLE_DATE_BEFORE_HOLIDAY',
]
for var in num_vars:
    lists_train[var + '_norm'] = lists_train[var].astype(float) / np.std(lists_train[var])
```

```python
visit_train_merged = merge(
    left=visit_train,
    right=coupon_detail_train.drop(['USER_ID_hash','I_DATE'], axis=1),
    on='PURCHASEID_hash',
    how='left',
)

visit_train_merged = merge(
    left=visit_train_merged,
    right=lists_train,
    on='COUPON_ID_hash',
    how='left',
)

visit_train_merged = merge(
    left=visit_train_merged,
    right=users,
    on='USER_ID_hash',
    how='left',
)
```

```python
# data structure
user_purchase = {}
for idx in coupon_detail_train.index:
    user_id = coupon_detail_train.ix[idx, 'USER_ID_hash']
    coupon_id = coupon_detail_train.ix[idx, 'COUPON_ID_hash']
    
    if user_purchase.get(user_id) is None:
        user_purchase[user_id] = []
    user_purchase[user_id].append(coupon_id)
```
